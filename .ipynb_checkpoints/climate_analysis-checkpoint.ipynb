# Climate Analysis and Exploration

```python
#Import all dependencies 
import sqlalchemy
from sqlalchemy.sql import label
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import pandas as pd
from sqlalchemy import Column, Integer, String, Numeric, Text, Float,Table,ForeignKey
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt

from datetime import datetime,timedelta,date
import datetime as datetime
```

```python
#Connection to engine 
engine = create_engine("sqlite:///hawaii.sqlite")
conn = engine.connect()
Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes.keys()
```

```python
#Functions for callings the columns
measurements = Base.classes.Measurements
stations = Base.classes.Stations
```

```python
#Creating a session
session = Session(engine)
```

# Precipitation Analysis 


```python
#Datetime to measure date from last year
year_ago = datetime.datetime.today() - datetime.timedelta(days = 365) 
str(year_ago)
```

```python
# Results of date and precitpiation dates for last year 
results = session.query(measurements.date, measurements.prcp).filter(measurements.date > year_ago).all()
results
```

```python
# Putting the results into a dataframe 
precipitation_df = pd.DataFrame(results, columns = ["Date", "Precipitation"])
precipitation_df.set_index("Date", inplace = True)
precipitation_df.head()
```

```python
#Plotting the data
plot_data = precipitation_df['Precipitation'].astype(float)
plot_data
```

```python
#Plot the results using the DataFrame `plot` method.
plot_data.plot(title="Precipitation over the last 12 months ",figsize=(10,5))
plt.show()
```

```python
plot_data.describe()
```

# Station Analysis 

```python
#Query into all stations
station_analysis = session.query(func.count(stations.station)).all()
station_analysis
```

```python
#Query to find the most active stations in descending order, Station USC00519281 has highest observations
active_station = session.query(func.count(measurements.station),measurements.station).\
group_by(measurements.station).order_by(func.count(measurements.station).desc()).all()


most_active = active_station [0][1]
num_observations = active_station [0][0]

print("The most active station is: %s" % most_active)
print("Number of observations for station %s: %s" % (most_active, num_observations))
```

```python
#Query to retrieve the last 12 months of temperature observation data (tobs).
year_ago = datetime.datetime.today() - datetime.timedelta(days = 365) 
str(year_ago)

#Filter by the station with the highest number of observations
tobs = session.query(measurements.tobs).filter(measurements.date > year_ago).\
    filter(measurements.station == "USC00519281").all()
```

```python
#Tobs Results in a dataframe
tobs_df = pd.DataFrame(tobs)
tobs_df.head()
```

```python
#Plot graph
frequency = pd.DataFrame(tobs)
tobs_df = frequency.astype(float)
tobs_df.plot.hist(bins = 12)
sns.set_style("darkgrid")
plt.show()
```
